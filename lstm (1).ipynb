# Time-Series Prediction with LSTM

## Problem Statement

The goal is to predict future values of a time-series dataset using an LSTM (Long Short-Term Memory) neural network. Specifically, given historical data on `torque` and `speed`, the model predicts the `speed` for future time steps based on past `torque` values.

## Code Explanation

### 1. **Importing Libraries**

```python
import pandas as pd

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt

# Function to load CSV files with error handling
def load_csv_with_error_handling(file_path):
    try:
        # Load CSV file and skip bad lines
        data = pd.read_csv(file_path, header=0, on_bad_lines='skip')
        return data
    except pd.errors.ParserError as e:
        print(f"Error loading CSV file {file_path}: {e}")
        return pd.DataFrame()  # Return an empty DataFrame on error

# Paths to your CSV files
train_csv_file_path = '/content/output_data.csv'  # Replace with your training CSV file path
test_csv_file_path = '/content/test.csv'    # Replace with your testing CSV file path

# Load datasets
train_data = load_csv_with_error_handling(train_csv_file_path)
test_data = load_csv_with_error_handling(test_csv_file_path)

# Check if datasets are loaded properly
if train_data.empty or test_data.empty:
    raise ValueError("One or both CSV files could not be loaded properly.")

# Display the first few rows of the datasets
print("Training Data:")
print(train_data.head())
print("Testing Data:")
print(test_data.head())

# Drop any rows with non-numeric values in the 'speed' or 'torque' columns
train_data = train_data[pd.to_numeric(train_data['speed'], errors='coerce').notnull()]
test_data = test_data[pd.to_numeric(test_data['speed'], errors='coerce').notnull()]

# Convert columns to numeric
train_data['speed'] = pd.to_numeric(train_data['speed'])
train_data['torque'] = pd.to_numeric(train_data['torque'])
test_data['speed'] = pd.to_numeric(test_data['speed'])
test_data['torque'] = pd.to_numeric(test_data['torque'])

# Combine training and testing data for consistent scaling
combined_data = pd.concat([train_data, test_data])
scaler = MinMaxScaler(feature_range=(0, 1))

# Scale the features
scaled_features = scaler.fit_transform(combined_data[['speed', 'torque']])

# Prepare data for LSTM
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length, 1]  # Using 'torque' as feature
        y = data[i + seq_length, 0]  # Predicting 'speed'
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

sequence_length = 10  # Number of timesteps

# Prepare training sequences
train_scaled = scaled_features[:len(train_data)]
X_train, y_train = create_sequences(train_scaled, sequence_length)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

# Prepare testing sequences
test_scaled = scaled_features[len(train_data):]
X_test, y_test = create_sequences(test_scaled, sequence_length)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(sequence_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=20, verbose=1, validation_split=0.2)

# Evaluate the model
y_pred = model.predict(X_test)
y_test_inv = scaler.inverse_transform(np.concatenate((np.reshape(y_test, (-1, 1)), np.zeros((len(y_test), 1))), axis=1))[:, 0]
y_pred_inv = scaler.inverse_transform(np.concatenate((np.reshape(y_pred, (-1, 1)), np.zeros((len(y_pred), 1))), axis=1))[:, 0]

mse = mean_squared_error(y_test_inv, y_pred_inv)
rmse = math.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")

# Plot training history
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
````
